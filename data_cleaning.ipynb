```python
#Set up replacement dictionary
replace_type = {"type": {"tvSeries": "tvShow",
                        "tvEpisode": "tvShow", 
                        "tvSpecial": "movie",
                        "tvMiniSeries": "tvShow",
                        "tvMovie": "movie",
                        "video": "movie",
                        "short": "movie",
                        "tvShort": "tvShow"
                         },  
               }
```

```python
#Replace type so that all are movie or tvshow
netflix1_df_replacetype = netflix1_df_droptype.replace(replace_type)
netflix1_df_replacetype
```

```python
#Check values in type
netflix1_df_replacetype.value_counts(['type'])
```

```python
#Check all genres
all_genres = dict(netflix1_df_replacetype['genres'].value_counts())
print(all_genres)
```

```python
#Split genres into 3 columns
netflix1_df_replacetype[['Genre1','Genre2','Genre3']] = netflix1_df_replacetype["genres"].str.split(',', expand=True)
netflix1_df_replacetype.head() 
```

```python
netflix1_df_clean = netflix1_df_replacetype.drop(columns=['genres'])
netflix1_df_clean
```

```python
#Set up rating replacement dictionary
replace_rating = {"Movie Rating": {"TV-MA": "adult",
                        "TV-14": "teen", 
                        "R": "adult",
                        "TV-PG": "children",
                        "PG-13": "teen",
                        "PG": "children",        
                        "TV-Y7": "children",
                        "TV-Y": "children",
                        "TV-G": "children",
                        "NR": "adult",
                        "G": "children",
                        "TV-Y7-FV": "children",
                        "NC-17": "adult",
                        "UR": "adult",
                         },  
               }
```

```python
#Replace rating so that all are binned into adult, children, or teen 
netflix2_df_replacerating = netflix2_df_dropped_data.replace(replace_rating)
netflix2_df_replacerating
```

```python
#Check all durations
all_types = dict(netflix2_df_replacerating['duration'].value_counts())
print(all_types)
```

```python
#Drop country, date_added, and duration
netflix2_df_dropped_2 = netflix2_df_replacerating.drop(columns=['country','date_added','duration'])
netflix2_df_dropped_2.head()
```
